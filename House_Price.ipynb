```python
acc_to_6th = 0
acc_to_5th = 0
acc_to_4th = 0
acc_to_3rd = 0
acc_to_2nd = 0
acc_to_1st = 0

for pred in range(y_pred.shape[0]):
  str_of_int_pred = str(y_pred[pred])
  str_of_int = str(y_train[pred])

  if len(str_of_int_pred.split(".")[0]) == 6:
    if str_of_int_pred[0] == str_of_int[0]:
      acc_to_6th += 1

      if str_of_int_pred[1] == str_of_int[1]:
        acc_to_5th += 1

        if str_of_int_pred[2] == str_of_int[2]:
          acc_to_4th += 1

          if str_of_int_pred[3] == str_of_int[3]:
            acc_to_3rd += 1

            if str_of_int_pred[4] == str_of_int[4]:
              acc_to_2nd += 1

              if str_of_int_pred[5] == str_of_int[5]:
                acc_to_1st += 1
    # else:
      # print(f"{str_of_int_pred}  {str_of_int}")

  if len(str_of_int_pred.split(".")[0]) == 5:
    if str_of_int_pred[0] == str_of_int[0]:
      acc_to_6th += 1

      if str_of_int_pred[1] == str_of_int[1]:
        acc_to_5th += 1

        if str_of_int_pred[2] == str_of_int[2]:
          acc_to_4th += 1

          if str_of_int_pred[4] == str_of_int[4]:
              acc_to_3rd += 1

              if str_of_int_pred[5] == str_of_int[5]:
                acc_to_2nd += 1
```

```python
print(f"Accuracy to the first digit's : {(acc_to_6th / y_pred.shape[0])*100:.5f}%")
print(f"Accuracy to the second digit : {(acc_to_5th / y_pred.shape[0])*100:.5f}%")
print(f"Accuracy to the third digit : {(acc_to_4th / y_pred.shape[0])*100:.5f}%")
print(f"Accuracy to the fourth digit : {(acc_to_3rd / y_pred.shape[0])*100:.5f}%")
print(f"Accuracy to the fifth digit : {(acc_to_2nd / y_pred.shape[0])*100:.5f}%")
print(f"Accuracy to the sixth digit : {(acc_to_1st / y_pred.shape[0])*100:.5f}%")
```
